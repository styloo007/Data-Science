```python
data.drop(columns=['SeniorCitizen', 'tenure', 'MonthlyCharges'], inplace=True)
```

```python
encoded_data=pd.get_dummies(data)
```

```python
encoded_data.columns
```

```python
X= encoded_data.drop(columns='Churn_Yes')

Y=encoded_data['Churn_Yes']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

model=Lasso(alpha=0.1)

model.fit(X_train, Y_train)

```

```python
predictions=model.predict(X_test)

```

```python
# Convert Predictions to Binary Values

predictions_bin=(predictions>=0.5).astype(int)
```

```python
print(classification_report(Y_test, predictions_bin))
```

```python
print(accuracy_score(Y_test, predictions_bin))
```

```python
conf=confusion_matrix(Y_test, predictions_bin)
print(conf)
```
